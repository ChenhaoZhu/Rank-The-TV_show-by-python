```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import urllib
from bs4 import BeautifulSoup
print('packages imported sucessfully !')
```

```python
url = 'https://www.meijutt.tv/alltop_hit.html'
headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.87 Safari/537.36 SLBrowser/6.0.1.4221'
}
response = requests.get(url = url, headers = headers)
page = response.content.decode('gbk')  # 找到含有所有信息的源代码，如果一开始utf-8不行，可以试试gbk
print('Original Code has been Stolen Sucessfully !')
data = BeautifulSoup(page, 'lxml')
print('Transfer original code into BeautifulSoup Object Sucessfully !')

score1 = data.select('strong')
score2 = data.select('span')
s = []
count = 0
for aft in score2 :
    if '.' in aft.text :
        pre = score1[count].text
        score = float(pre + aft.text)
        s.append(score)
print('Socres have been saved Sucessfully !')

name = data.select('h5')
dic = {}
c = 0
for each in name :
    if 'href' not in str(each) :
        ca = str(each.text)
        dic[ca] = []
        print(ca + ', 类别的电影已被建立')
    else :
        dic[ca].append([str(each.text), s[c]])
        print(ca + ' 类别电影增加 : ' + str(each.text) + ', 评分 : ' + str(s[c]))


    
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
fig = plt.figure()
plt.style.use('ggplot')
ax = fig.add_subplot(1,1,1)
ax.hist(dic['魔幻/科幻'])
plt.show()
```
